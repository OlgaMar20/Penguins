Penguins are flightless seabirds that mainly inhabit the southern hemisphere, although some species are also found in regions further north. They are famous for their peculiar appearance, with streamlined bodies, modified wings in the form of fins and a characteristic black and white plumage pattern. Penguins are known for their ability to swim and dive, which makes them excellent hunters of fish and other marine prey.

![Ping칲inos](1.jpeg)


#### **Distribution and Habitat**
Penguins are distributed in various coastal and oceanic regions of the world, with most species concentrated in the southern hemisphere. The areas best known for hosting significant penguin populations include Antarctica, sub-Antarctic islands such as the Falkland Islands and South Georgia, as well as the coasts of South Africa, Australia, New Zealand and South America. They prefer coastal habitats with access to cold waters and rich in food, such as krill, fish and squid.

![Penguins](2.jpg)

#### **Physical Characteristics**
Penguins have unique adaptations for life in the aquatic environment. They have fusiform bodies that allow them to glide easily through water with minimal resistance. Their wings, transformed into fins, provide them with momentum underwater, while their webbed feet act as rudders. Their dense and oily plumage helps them to stay warm and isolated from the cold of the water. In addition, they have special glands that secrete an oil that cover their plumage, waterproofing it and providing them with buoyancy.

![Penguins](3.jpg)

#### **Behavior and Life Cycle**
Penguins are social animals that form colonies to reproduce and raise their young. Many species exhibit ritualized courtship and mating behavior. After incubation, penguins lay one or two eggs, depending on the species. Both parents share responsibilities in the incubation and feeding of the chicks. Some species migrate long distances in search of food, while others remain in their breeding areas throughout the year.

![Penguins](4.jpeg)

#### **Featured Penguin Species**
1. **Adelie Penguin (Pygoscelis adeliae)**
The Adelie penguin is one of the most common and widespread species of penguins. It is found mainly in the Antarctic region and on the Subantarctic islands. They are medium-sized, with a black plumage on the back and head, and a white belly. They are known for their curious personality and for being excellent swimmers and divers.

![Penguins](5.jpeg)

2. **Chinstrap Penguin (Pygoscelis antarctica)**
The Chinstrap penguin is a medium-sized and common species in the Antarctic region. It stands out for the white crescent-shaped patch that wraps around its chin, hence its name. It prefers coastal habitats and forms large colonies during the breeding season. They are excellent swimmers and divers, feeding mainly on fish and krill. Its population is considered stable, but it faces risks from climate change and human activity in its habitat.

![Penguins](10.jpeg)

3. **Gentoo Penguin (Pygoscelis papua)**
The Gentoo penguin is recognizable by its black plumage on the back, head and wings, and a white head patch behind the eye. They are the third largest species of penguins and are found on several Subantarctic islands and on the Antarctic peninsula. They are agile swimmers and feed on a variety of marine prey, including fish, squid and krill.

![Penguins](6.jpg)

4. **Emperor Penguin (Aptenodytes forsteri)**
The Emperor penguin is the largest and most emblematic species of penguins. It is found exclusively in Antarctica and is known for its ability to withstand extremely cold and harsh conditions. They have a white plumage on the belly and black on the back and head, with striking orange spots on the neck and upper chest. Emperor penguins are known for their long treks to reach their breeding areas and their ability to raise their young in the harshest conditions of the Antarctic winter.

![Penguins](7.jpg)

#### **Conservation and Threats**
Many penguin populations face significant threats, including habitat loss, pollution, overfishing and climate change. The decline of prey populations, such as krill, can also negatively affect the penguins that depend on them for food. Several penguin species are listed as endangered or vulnerable according to the IUCN Red List, highlighting the importance of conservation efforts to protect their habitats and ensure their long-term survival.

```python
# Importamos las librer칤as
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Cargamos el dataset 
penguins=sns.load_dataset('penguins')
```

```python
# Crear una funci칩n para generar datos aleatorios para un ping칲ino Emperador
def generar_datos_emperador():
    species = 'Emperor'
    island = 'Emperor_Island'  # Asignar un valor espec칤fico para la isla
    bill_length_mm = np.random.normal(45, 5)  # Longitud del pico (media=45, desviaci칩n est치ndar=5)
    bill_depth_mm = np.random.normal(15, 2)   # Profundidad del pico (media=15, desviaci칩n est치ndar=2)
    flipper_length_mm = np.random.normal(220, 20)  # Longitud de la aleta (media=220, desviaci칩n est치ndar=20)
    body_mass_g = np.random.normal(6500, 500)  # Peso del cuerpo (media=6500, desviaci칩n est치ndar=500)
    sex = np.random.choice(['Male', 'Female', np.nan], p=[0.5, 0.5, 0.0])  # Sexo (50% Male, 50% Female, 0% NaN)
    return [species, island, bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g, sex]

# Generar 172 registros para ping칲inos Emperador
emperador_registros = [generar_datos_emperador() for _ in range(172)]

# Convertir la lista de registros en un DataFrame
df_emperador = pd.DataFrame(emperador_registros, columns=['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex'])

# Concatenar el DataFrame de Emperador con el dataset de ping칲inos original
penguins = pd.concat([penguins, df_emperador], ignore_index=True)

# Mostrar los primeros registros del dataset actualizado
penguins.head()
```

```python
# Crear el gr치fico de conteo
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=penguins, palette='pastel')
plt.title('Distribuci칩n de Especies de Ping칲inos')
plt.xlabel('Especie')
plt.ylabel('Conteo')
plt.xticks(rotation=45)  # Rotar etiquetas en el eje x para mayor claridad
plt.show()
```

```python
# Diagrama de caja para comparar la masa corporal de los ping칲inos en funci칩n de su especie y sexo

# Crear el diagrama de caja
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='body_mass_g', hue='sex', data=penguins)
plt.title('Diagrama de Caja: Masa Corporal de Ping칲inos por Especie y Sexo')
plt.xlabel('Especie')
plt.ylabel('Masa Corporal (g)')
plt.legend(title='Sexo')
plt.show()
```

```python
# Crear el diagrama de dispersi칩n
plt.figure(figsize=(10, 6))
sns.scatterplot(x='flipper_length_mm', y='body_mass_g', hue='species', data=penguins)
plt.title('Diagrama de Dispersi칩n: Longitud de Aletas vs. Masa Corporal por Especie')
plt.xlabel('Longitud de Aletas (mm)')
plt.ylabel('Masa Corporal (g)')
plt.legend(title='Especie')
plt.show()
```

```python
# Crear la gr치fica de pares
sns.pairplot(penguins, hue='species')
plt.suptitle('Gr치fica de Pares: Exploraci칩n de Relaciones entre Variables Num칠ricas', y=1.02)
plt.show()
```
